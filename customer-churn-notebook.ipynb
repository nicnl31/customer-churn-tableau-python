```python
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{:.1f}%\n({v:d})'.format(pct, v=val)
    return my_format
```

```python
churn_cats = df1["Customer Status"].value_counts()
churn_cats
```

## Stayed vs Churned

```python
plt.pie(churn_cats,labels = churn_cats.index, autopct=autopct_format(churn_cats))
plt.title("Ratio of Stayed vs. Churned", size=16, weight="bold")
plt.savefig("py_stayed-vs-churned.png")
plt.show()
```

## Age and gender distribution by cohort

```python
age_bins = [18, 25]
age_bins.extend([i for i in range(35, 66, 10)])
age_bins.extend([80])
```

```python
df_stayed = df1[df1["Customer Status"] == "Stayed"]
df_churned = df1[df1["Customer Status"] == "Churned"]
```

```python
stayed_age_bins = pd.cut(df_stayed['Age'], bins=age_bins).value_counts(normalize=True).sort_index()
churned_age_bins = pd.cut(df_churned['Age'], bins=age_bins).value_counts(normalize=True).sort_index() * -1
```

```python
sns.barplot(x=stayed_age_bins, y=stayed_age_bins.index, color="#4aaf8d", label="Stayed")
sns.barplot(x=churned_age_bins, y=churned_age_bins.index, color="#ee8c5f", label="Churned")

plt.xticks([-0.2, -0.15, -0.10, -0.05, 0.0, 0.05, 0.10, 0.15, 0.2], ['20', '15', '10', '5', '0', '5', '10', '15', '20'])
plt.xlabel("% within Churned/Stayed Cohort")
plt.ylabel("Age band")
plt.title("Percentage Distribution of Age by Churned/Stayed Cohort", size=16, weight="bold")
plt.axvline(color="#000000")
plt.legend(loc="upper left", frameon=True)
sns.despine(left=True, bottom=True)
plt.savefig("py_age-dist-by-cohort.png")
plt.show()
```

```python
ax = sns.countplot(data=df1, x="Gender", hue="Customer Status")
plt.ylabel("Count")
plt.title("Gender Count by Churned/Stayed Cohort", size=16, weight="bold")
plt.legend(loc="upper right", frameon=True)
sns.despine(left=True, bottom=True)
plt.savefig("py_gender-count.png")
plt.show()
```

```python
tenure_monthlycharge_group = df1[["Tenure in Months", "Monthly Charge", "Customer Status"]].groupby(["Tenure in Months", "Customer Status"]).mean()
tenure_totalcharges_group = df1[["Tenure in Months", "Total Charges", "Customer Status"]].groupby(["Tenure in Months", "Customer Status"]).mean()
tenure_totalextradatacharges_group = df1[["Tenure in Months", "Total Extra Data Charges", "Customer Status"]].groupby(["Tenure in Months", "Customer Status"]).mean()
tenure_totallongdistcharges_group = df1[["Tenure in Months", "Total Long Distance Charges", "Customer Status"]].groupby(["Tenure in Months", "Customer Status"]).mean()

charges_datas = [
    tenure_monthlycharge_group, 
    tenure_totalcharges_group,
    tenure_totalextradatacharges_group,
    tenure_totallongdistcharges_group
]

charges_names = [
    "Monthly Charge", 
    "Total Charges", 
    "Total Extra Data Charges", 
    "Total Long Distance Charges"
]
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams["figure.dpi"] = 200

# Churned, Stayed
palette = ["#ee8c5f", "#4aaf8d"]
sns.set_palette(sns.color_palette(palette))

rows, cols = 2, 2
fig = plt.figure()
for i in range(rows*cols):
    ax = fig.add_subplot(rows, cols, i+1)
    sns.scatterplot(data=charges_datas[i], x="Tenure in Months", y=charges_names[i], 
                    hue="Customer Status", style="Customer Status", markers=['X', 'o'])
    ax.set_xlabel(f"Tenure in Months")
    ax.set_ylabel(f"Average {charges_names[i]}")
    ax.set_title(f"Subplot {i+1}: Average {charges_names[i]}", style='italic')
    ax.legend(frameon=True)

fig.suptitle("Average Charge Categories by Customer Tenure in Months",size=16, weight="bold")    
fig.tight_layout()
plt.savefig("py_tenure-vs-charge.png")
fig.show()
```

```python
tenure_referral_group_churned = df1[df1["Customer Status"]=="Churned"][["Tenure in Months", "Number of Referrals"]].groupby(["Tenure in Months"]).mean()
tenure_referral_group_stayed = df1[df1["Customer Status"]=="Stayed"][["Tenure in Months", "Number of Referrals"]].groupby(["Tenure in Months"]).mean()
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.rcParams["figure.dpi"] = 200

sns.regplot(data=tenure_referral_group_churned, x=tenure_referral_group_churned.index, y="Number of Referrals", label="Churned", marker='X')
sns.regplot(data=tenure_referral_group_stayed, x=tenure_referral_group_stayed.index, y="Number of Referrals", label="Stayed", marker='o')
plt.ylabel("Average Number of Referrals")
plt.title("Tenure in Months against Average Number of Referrals by Customer\n Status, plus Trend Lines with 95% Confidence Bands",
         size=16, weight="bold")
plt.legend(frameon=True)
plt.savefig("py_tenure-vs-referrals.png")
plt.show()
```

Churners are less enthusiastic than stayers in terms of the average number of referrals as they stay longer, as reflected in both the relative position of the clusters as well as the slope of the individual trend lines.
